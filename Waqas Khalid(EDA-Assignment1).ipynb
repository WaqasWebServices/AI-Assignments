```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
#heat map

sns.heatmap(df[['Confirmed', 'Deaths', 'Recovered']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#bivariate analysis
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Confirmed', y='Deaths', data=df, alpha=0.5)
plt.title('Confirmed vs Deaths')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.show()
```

```python
#multivariate analysis
sns.violinplot(df[['Confirmed', 'Deaths', 'Recovered']])
plt.suptitle('Multivariate Analysis', y=1.02)
plt.show()
```
