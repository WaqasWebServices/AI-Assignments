```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 9, 2, 7, 14])
y = np.array([12, 6, 6, 8, 4]) 
```

```python
def gradient_descent(X, y, lr=0.01, epochs=1000):
    m = len(y)
    X = X.reshape(-1, 1)
    y = y.reshape(-1, 1)
    theta0 = 0
    theta1 = 0
    for _ in range(epochs):
        y_pred = theta0 + theta1 * X
        d_theta0 = (1/m) * np.sum(y_pred - y)
        d_theta1 = (1/m) * np.sum((y_pred - y) * X)
        theta0 -= lr * d_theta0
        theta1 -= lr * d_theta1
    return theta0, theta1

theta0, theta1 = gradient_descent(X, y)
print("Intercept:", theta0)
print("Slope:", theta1)
```

```python
y_pred = theta0 + theta1 * X

plt.scatter(X, y, color='violet', label='Actual Data')
plt.plot(X, y_pred, color='maroon', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Univariate Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
